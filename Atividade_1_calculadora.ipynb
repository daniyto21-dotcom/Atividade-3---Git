```python
#Calculadora Interativa
print('Bem-vindo à calculadora interartiva')
print('Selecione a operação que deseja realizar')

while True:
  print('Escolha uma operação')
  print('1 - Somar(+)')
  print('2 - Subtrair(-)')
  print('3 - Multiplicar(*)')
  print('4 - Dividir(/)')
  print('5 - Sair do Programa')

  #recebendo a opção do usuário
  opção = input("Digite o número da sua opção(1-5)")

  #Condicional se o usuário quiser sair do programa
  if opção == '5':
      print ('Obrigado por usar a calculadora!')
      break

  #Condicional para garantir que o usuário escolheu uma opção válida
  if opção in ['1', '2', '3', '4']:

  #Requisito 1 e 2: tranformar input para float (tranforma texto em número real)
      num1 = int(input('Digite o primeiro número (1-9):'))
      num2 = int(input('Digite o segundo número (1-9):'))

  #Requisitos 3 e 4: condicionais aplicando as 4 operações matemáticas

      if opção == '1':
               resultado = num1 + num2
               operação_nome = 'somar'
      elif opção == '2':
              resultado = num1 - num2
              operação_nome = 'subtrair'
      elif opção == '3':
               resultado = (num1 * num2)
               operação_nome = 'multiplicar'
      elif opção == '4':
               resultado = (num1 / num2)
               operação_nome = 'dividir'

      print("Resultado =", resultado)




```
